```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
#file_path=

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv("Resources/purchase_data.csv")
players=purchase_data.iloc[:,1]
players_list=players.unique()
player_count=pd.DataFrame(data=players_list).count()
Total_Players=pd.DataFrame(data=player_count,columns=['Total Players'])
Total_Players
```

```python
#Run basic calculations to obtain number of unique items, average price, etc.
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame
```

```python
items=purchase_data.iloc[:,4]
unique_items=items.unique()
price=purchase_data.iloc[:,6]
total_price=pd.DataFrame(data=price).sum()
purchase_id=purchase_data.iloc[:,0]
average_price=int(total_price)/len(items)
Purchasing_Analysis=pd.DataFrame({'Number of Unique items':[len(unique_items)],
                                'Average Price':[float(average_price)],
                                'Number of purchases':[len(purchase_id)],
                                'Total Revenue':[float(total_price)]})
pd.options.display.float_format = '${:,.2f}'.format
Purchasing_Analysis
```

```python
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

#set "Gender"=="Male","Female","Other" and count then divided by len(list)
male_purchases=purchase_data.loc[(purchase_data["Gender"]=="Male"),:]
female_purchases=purchase_data.loc[(purchase_data["Gender"]=="Female"),:]
other_purchases=purchase_data.loc[(purchase_data["Gender"]=="Other / Non-Disclosed"),:]

#find total amount of each gender
total_males=male_purchases["SN"].unique()
total_females=female_purchases["SN"].unique()
total_others=other_purchases["SN"].unique()

#total of how many players
total_unique_players=len(total_males)+len(total_females)+len(total_others)

#percentage by gender
male_percent=(len(total_males)/(total_unique_players)*100)
female_percent=(len(total_females)/(total_unique_players)*100)
other_percent=(len(total_others)/(total_unique_players)*100)

#make into DataFrame
Gender_percentage_chart=pd.DataFrame({
    "Total":[len(total_males),len(total_females),len(total_others)],
    "Percentage of Players":[male_percent,female_percent,other_percent]
})
#add formatting to get it to two decimals
pd.options.display.float_format = '{:,.2f}%'.format
#add row formatting to get names to change
Gender_percentage_chart.rename(index={0:'Male',1:'Female',2:'Other / Non-Disclosed'}, inplace=True)

Gender_percentage_chart
```

```python
#Run basic calculations to obtain {purchase count, 
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

Purchasing_Analysis=pd.DataFrame({
    "Purchase Count":[len(male_purchases),len(female_purchases),len(other_purchases)],
    "Average Purchase Price":[((male_purchases['Price'].sum())/len(male_purchases)),((female_purchases['Price'].sum())/len(female_purchases)),((other_purchases['Price'].sum())/len(other_purchases))],
    "Total Purchase Value":[(male_purchases['Price'].sum()),(female_purchases['Price'].sum()),(other_purchases['Price'].sum())],
    "Average Total Purchase per Person":[((male_purchases['Price'].sum())/len(total_males)),((female_purchases['Price'].sum())/len(total_females)),((other_purchases['Price'].sum())/len(total_others))]
})
#add formatting to get it to two decimals
pd.options.display.float_format = '${:,.2f}'.format
#add row formatting to get names to change
Purchasing_Analysis.rename(index={0:'Male',1:'Female',2:'Other / Non-Disclosed'}, inplace=True)
Purchasing_Analysis.reindex(["Female", "Male", "Other / Non-Disclosed"])
```

```python
age = [0,9,14,19,24,29,34,39,99]
age_bracket = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Add column of bins based on Age
purchase_data["Age Group"]= pd.cut(purchase_data["Age"],age, labels=age_bracket)

age_demographics=purchase_data.groupby("Age Group")["SN"].nunique()

#len(age_demographics)

#age_demographics['unique'].count()
Demographics=pd.DataFrame(age_demographics,index=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'])
#Demographics=pd.DataFrame(age_demographics) ****contains index name
Demographics['percentage']=(Demographics['SN']/576)*100
pd.options.display.float_format = '{:,.2f}%'.format
Demographics.rename(columns={"SN": "Total Count", "percentage": "Percentage of Players"})
```

```python
#total count vs total sum
age_purchases=purchase_data.groupby("Age Group")["SN"].count()
total_purchases_by_age=purchase_data.groupby("Age Group")["Price"].sum()
#index needed now
purchases_by_age=pd.DataFrame(age_purchases,index=["10-14", "15-19", "20-24","25-29", "30-34", "35-39","40+","<10"])
purchases_by_age["Average Purchase Price"]=(total_purchases_by_age)/purchases_by_age["SN"]
purchases_by_age["Total Purchase Value"]=total_purchases_by_age

purchases_by_age["Avg Total Purchase per Person"]=(purchases_by_age["Total Purchase Value"])/age_demographics
pd.options.display.float_format = '${:,.2f}'.format
#**doesntwork***.reindex(["10-14", "15-19", "20-24","25-29", "30-34", "35-39","40+","<10"])
purchases_by_age.rename(columns={"SN": "Purchase Count"})
```

```python
big_spenders=purchase_data.groupby("SN")["Price"].sum()
total_purchases_by_person=purchase_data.groupby("SN")["Purchase ID"].count()
top_big_spenders=pd.DataFrame(big_spenders).sort_values(["Price"], ascending=False)
top_big_spenders["Purchase Count"]=total_purchases_by_person
top_big_spenders["Average Purchase Price"]=(top_big_spenders["Price"])/(top_big_spenders["Purchase Count"])
#re-order columns
top_big_spenders=top_big_spenders[["Purchase Count","Average Purchase Price","Price"]]
top_big_spenders.rename(columns={"Price": "Total Purchase Value"}).head()
```

```python
purchase_data=purchase_data.groupby(['Item ID','Item Name'])['Price'].agg(['count','sum','mean']).sort_values(['sum'], ascending= False).reset_index()
dude=purchase_data
most_profitable_item=purchase_data.set_index(['Item ID', 'Item Name'])
most_profitable_item=most_profitable_item[["count","mean","sum"]]
most_profitable_item.rename(columns={"count": "Purchase Count", "mean": "Item Price","sum": "Total Purchase Value"}).head()
most_bought_item=most_profitable_item.sort_values(by=['count','Item ID'],ascending=False)
most_bought_item.rename(columns={"count": "Purchase Count", "mean": "Item Price","sum": "Total Purchase Value"}).head()
```
